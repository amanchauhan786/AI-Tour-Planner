````python
import requests
import pandas as pd
from geopy.geocoders import Nominatim

# Replace with your OpenRouteService API Key
ORS_API_KEY = "5b3ce3597851110001cf62483783c9a0c7114a63a761cfc79a39a650"

def get_coordinates(location_name):
    """
    Converts a place name into latitude and longitude using Geopy.
    Example: "New York, USA" → (40.7128, -74.0060)
    """
    geolocator = Nominatim(user_agent="geoapi")
    location = geolocator.geocode(location_name)

    if location:
        return f"{location.longitude},{location.latitude}"
    else:
        return None

def fetch_route(start_location, end_location, mode="driving-car"):
    """
    Fetches route details from OpenRouteService API.
    Supports modes: "driving-car", "cycling-regular", "foot-walking", "public-transport".
    """
    start_coords = get_coordinates(start_location)
    end_coords = get_coordinates(end_location)

    if not start_coords or not end_coords:
        return {"Error": "Invalid location names. Try using a more specific location."}

    url = f"https://api.openrouteservice.org/v2/directions/{mode}"
    headers = {"Authorization": ORS_API_KEY}
    params = {"start": start_coords, "end": end_coords}

    response = requests.get(url, headers=headers, params=params)
    return response.json()

def process_route_data(route_json):
    """
    Extracts travel distance, duration, and step-by-step directions.
    """
    try:
        route = route_json["routes"][0]
        distance_km = round(route["summary"]["distance"] / 1000, 2)  # Convert meters to km
        duration_min = round(route["summary"]["duration"] / 60, 2)  # Convert seconds to minutes
        steps = []

        for step in route["segments"][0]["steps"]:
            steps.append(f"{step['instruction']} - {step['distance']} meters")

        return {
            "Total Distance (km)": distance_km,
            "Total Duration (min)": duration_min,
            "Step-by-Step Directions": steps
        }

    except Exception as e:
        return {"Error": str(e)}

def display_route_details(route_details):
    """
    Displays the route details in a readable format.
    """
    print(f"🚗 Total Distance: {route_details['Total Distance (km)']} km")
    print(f"⏳ Estimated Travel Time: {route_details['Total Duration (min)']} minutes")

    print("\n📍 Step-by-Step Directions:")
    for i, step in enumerate(route_details["Step-by-Step Directions"], 1):
        print(f"{i}. {step}")

    return pd.DataFrame(route_details["Step-by-Step Directions"], columns=["Navigation Steps"])

# 🚀 User Input for Home & Destination
home = input("Enter your starting location (e.g., 'New York, USA'): ")
destination = input("Enter your destination (e.g., 'Los Angeles, USA'): ")
travel_mode = input("Enter travel mode (driving-car, cycling-regular, foot-walking, public-transport): ")

# 🗺 Fetch and display route details
route_data = fetch_route(home, destination, travel_mode)
route_details = process_route_data(route_data)

# 📝 Show formatted output
if "Error" in route_details:
    print("❌ Error:", route_details["Error"])
else:
    df_route = display_route_details(route_details)
    print("\n📊 Route Details in DataFrame:")
    print(df_route.head())
````

````python
import requests
import folium
import openrouteservice
from geopy.geocoders import Nominatim

# Replace with your OpenRouteService API Key
ORS_API_KEY = "5b3ce3597851110001cf62483783c9a0c7114a63a761cfc79a39a650"

# Initialize OpenRouteService client
ors_client = openrouteservice.Client(key=ORS_API_KEY)

def get_coordinates(location_name):
    """
    Converts a location name into latitude and longitude using Geopy.
    """
    geolocator = Nominatim(user_agent="geoapi")
    location = geolocator.geocode(location_name)

    if location:
        return (location.latitude, location.longitude)
    else:
        return None

def get_route(start_location, end_location, mode="driving-car"):
    """
    Fetches route coordinates between two locations using OpenRouteService.
    """
    start_coords = get_coordinates(start_location)
    end_coords = get_coordinates(end_location)

    if not start_coords or not end_coords:
        return None, "Invalid locations. Please try again."

    route = ors_client.directions(
        coordinates=[(start_coords[1], start_coords[0]), (end_coords[1], end_coords[0])],
        profile=mode,
        format="geojson"
    )

    return route, start_coords, end_coords

def plot_route(route, start_coords, end_coords):
    """
    Plots the route on an interactive Folium map.
    """
    m = folium.Map(location=start_coords, zoom_start=12)

    # Add markers for start and end locations
    folium.Marker(start_coords, tooltip="Start", icon=folium.Icon(color="green")).add_to(m)
    folium.Marker(end_coords, tooltip="Destination", icon=folium.Icon(color="red")).add_to(m)

    # Extract route coordinates
    route_coords = [(point[1], point[0]) for point in route['features'][0]['geometry']['coordinates']]

    # Draw the route on the map
    folium.PolyLine(route_coords, color="blue", weight=5, opacity=0.7).add_to(m)

    return m

# 🚀 User Input for Home & Destination
home = input("Enter your starting location (e.g., 'New York, USA'): ")
destination = input("Enter your destination (e.g., 'Los Angeles, USA'): ")
travel_mode = input("Enter travel mode (driving-car, cycling-regular, foot-walking, public-transport): ")

# 🗺 Fetch and visualize the route
route_data, start_coords, end_coords = get_route(home, destination, travel_mode)

if route_data:
    m = plot_route(route_data, start_coords, end_coords)
    m.save("route_map.html")
    print("✅ Route map saved! Open 'route_map.html' to view the map.")
else:
    print("❌ Error fetching route. Please check the locations and try again.")
````

````python

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def scrape_oyo_hotels(city):
    url = f"https://www.oyorooms.com/hotels-in-{city}/"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
    }

    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        return f"Failed to fetch data: {response.status_code}"

    soup = BeautifulSoup(response.text, "html.parser")
    hotels = []

    for hotel in soup.find_all("div", class_="hotelCardListing"):
        name = hotel.find("h3", class_="listingHotelDescription__hotelName")
        price = hotel.find("span", class_="listingPrice__finalPrice")
        rating = hotel.find("span", class_="hotelRating__ratingSummary")
        address = hotel.find("span", class_="listingHotelDescription__hotelAddress")
        room_type = hotel.find("span", class_="c-gold")  # This class might contain room type details
        link = hotel.find("a")

        hotel_name = name.text.strip() if name else "N/A"
        hotel_address = address.text.strip() if address else "N/A"
        room_type_text = room_type.text.strip() if room_type else "Standard"

        # **Append "Hotel Name + City" for accurate weather search**
        location_query = f"{hotel_name} {city}"

        hotel_data = {
            "Hotel Name": hotel_name,
            "Price": price.text.strip() if price else "N/A",
            "Rating": rating.text.strip() if rating else "No rating",
            "Address": hotel_address,
            "Room Type": room_type_text,
            "Link": "https://www.oyorooms.com" + link["href"] if link else "N/A"
        }

        # Fetch weather using hotel name + city
        weather_data = scrape_weather(location_query)
        hotel_data.update(weather_data)  # Merge weather details into hotel_data

        hotels.append(hotel_data)

        # Add a delay to prevent Google blocking requests
        time.sleep(2)

    df = pd.DataFrame(hotels)
    df.to_csv(f"oyo_hotels_{city}_with_weather.csv", index=False)  # Save data as CSV
    return df

def scrape_weather(location_query):
    """Fetches weather data using 'Hotel Name + City'."""
    search_query = location_query.replace(" ", "+")
    weather_url = f"https://www.google.com/search?q=weather+{search_query}"

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
    }

    response = requests.get(weather_url, headers=headers)
    if response.status_code != 200:
        return {"Temperature": "N/A", "Condition": "N/A", "Humidity": "N/A", "Wind Speed": "N/A", "Precipitation": "N/A"}

    soup = BeautifulSoup(response.text, "html.parser")

    temperature = soup.find("span", class_="wob_t")
    condition = soup.find("div", class_="wob_dcp")
    humidity = soup.find("span", text="Humidity")
    wind_speed = soup.find("span", text="Wind")
    precipitation = soup.find("span", text="Precipitation")

    return {
        "Temperature": temperature.text.strip() if temperature else "N/A",
        "Condition": condition.text.strip() if condition else "N/A",
        "Humidity": humidity.find_next_sibling("span").text.strip() if humidity else "N/A",
        "Wind Speed": wind_speed.find_next_sibling("span").text.strip() if wind_speed else "N/A",
        "Precipitation": precipitation.find_next_sibling("span").text.strip() if precipitation else "N/A"
    }

def recommend_hotels_by_room_type(city):
    # Load the saved CSV file
    file_name = f"oyo_hotels_{city}_with_weather.csv"
    df = pd.read_csv(file_name)

    # Convert "Rating" column to numeric, replace "No rating" with 0
    df["Rating"] = pd.to_numeric(df["Rating"], errors='coerce').fillna(0)

    # Convert "Price" column to numeric, extracting only numbers
    df["Price"] = df["Price"].str.replace("₹", "").str.replace(",", "").astype(float)

    # **Categorizing Hotels Based on Room Type**
    def categorize_room_type(room_text):
        room_text = room_text.lower()
        if "single" in room_text:
            return "Single Room"
        elif "double" in room_text or "deluxe" in room_text:
            return "Double Room"
        elif "family" in room_text or "suite" in room_text:
            return "Family Room"
        else:
            return "Standard Room"

    df["Room Category"] = df["Room Type"].apply(categorize_room_type)

    # **Sorting Criteria:**
    # 1. Higher Rating (Descending)
    # 2. Lower Price (Ascending)
    df_sorted = df.sort_values(by=["Rating", "Price"], ascending=[False, True])

    # **Selecting the Top Hotels per Room Type**
    top_single = df_sorted[df_sorted["Room Category"] == "Single Room"].head(2)
    top_double = df_sorted[df_sorted["Room Category"] == "Double Room"].head(2)
    top_family = df_sorted[df_sorted["Room Category"] == "Family Room"].head(2)

    # **Merging All Recommendations**
    top_hotels = pd.concat([top_single, top_double, top_family])

    # Display categorized recommendations
    print("\nTop Recommended Hotels in", city.capitalize(), "by Room Type:\n")
    print(top_hotels[["Hotel Name", "Price", "Rating", "Room Category", "Address", "Temperature", "Condition", "Link"]])

    return top_hotels

# Example Usage
city = "mumbai"  # Replace with any city name
top_hotels_by_room_type_df = recommend_hotels_by_room_type(city)
````
